### DataFrame 2
- score.csv

```python
#  
# ""  
score = pd.read_csv('data/score.csv', encoding='cp949', index_col='')
score
```

```python
score.sum()
#  ->  axis = 0 ( )
```

```python
#   
score.sum(axis=1)
```

1.    DataFrame ("" column)

```python
score[''] = score.sum(axis=1)
score
```

2.    "" column

```python
score[''] = score[['1', '2', '3', '4']].mean(axis=1)
score
```

```python
# 
score.loc['':'Web', '1':'4'].mean(axis=1)
```

3.    ""  

```python
score.loc[''] = score.mean(axis=0)
score
```

4.        
   - max()
   - min()

```python
# 1.        (max_score)
max_score = score.iloc[:5,:4].max(axis=1)
# 2.      (min_score)
# 3. max_score - min_score
```

```python
max_score = score[['1', '2', '3', '4']].max(axis=1)
min_score = score[['1', '2', '3', '4']].min(axis=1)
```

```python
print(max_score, min_score)
```

```python
stats = score.describe().T
stats['sum'] = score.sum(axis=1)
stats['var'] = score.var(axis=1)
stats['std'] = score.std(axis=1)
stats['median'] = score.median(axis=1)
stats
```
