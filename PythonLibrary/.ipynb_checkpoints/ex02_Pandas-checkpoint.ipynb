### DataFrame 인덱싱, 슬라이싱
- 인덱싱 : 인덱스 번호(이름)을 통해 원하는 데이터 하나를 추출
  - []
- 슬라이싱 : 범위를 통해 원하는 여러개의 데이터를 추출

```python
# 2020 데이터 추출
df["2020"]
# Series 형태 추출(1차원)
```

```python
# 표와같은 2차원 형태로 추출
df[["2020"]]
```

```python
data3 = {'2020':[9668465,3391946,2942828,1450062],
        '2010' :[10312545,3567910,2758296,1454636],
         '2005':[9762546,3512547,2517680,1456016]
        }
df3 = pd.DataFrame(data3,index = ['서울','부산','인천','광주'])
df3
```

```python
# 부산
df3.loc['부산']
```

```python
# df명.loc[범위]
# df명.iloc[행, 열]
```

- DataFrame 데이터를 추출할때 인덱서를 활용
  - loc : 컬럼명, 인덱스명
  - iloc : 인덱스번호
- 사용법
  - df명.loc[행범위, 열범위]
- 간단하게 컬럼이름을 통해서 열하나만 추출할때 열인덱싱사용
  - df명.[컬럼명] -> Series 형태로 출력

### DataFrame 실습2
- score.csv

```python
# 데이터 읽어오기
# "과목" 인덱스로 변경
score = pd.read_csv('data/score.csv', encoding='cp949', index_col='과목')
score
```

```python
score.sum()
# 축설정 -> 기본 axis = 0 (행데이터끼리 연산)
```

```python
# 과목별 합계 
score.sum(axis=1)
```

1. 과목별 합계를 구하여 DataFrame에 추가하기("합계" column추가)

```python
score['합계'] = score.sum(axis=1)
score
```

2. 과목별 평균을 구해서 "평균" column추가

```python
score['평균'] = score[['1반', '2반', '3반', '4반']].mean(axis=1)
score
```

```python
# 인덱싱사용
score.loc['파이썬':'Web', '1반':'4반'].mean(axis=1)
```

3. 반별 평균을 계산해서 "반평균" 행 추가하기

```python
score.loc['반평균'] = score.mean(axis=0)
score
```

4. 과목별로 가장 높은 점수와 가장 낮은 점수의 차이
   - max()
   - min()

```python
# 1. 과목별 비교를 통해 가장 높은 점수 구하기 (max_score)
max_score = score.iloc[:5,:4].max(axis=1)
# 2. 과목별 가장 낮은 점수 구하기 (min_score)
# 3. max_score - min_score
```

```python
max_score = score[['1반', '2반', '3반', '4반']].max(axis=1)
min_score = score[['1반', '2반', '3반', '4반']].min(axis=1)
```

```python
print(max_score, min_score)
```

```python
stats = score.describe().T
stats['sum'] = score.sum(axis=1)
stats['var'] = score.var(axis=1)
stats['std'] = score.std(axis=1)
stats['median'] = score.median(axis=1)
stats
```
