```python
import random

ran = random.randint(1, 10)

while True:
    num = int(input("  >> "))
    
    if num > ran:
        print(f"{num}   .")
    elif num < ran:
        print(f"{num}   .")
    else:
        print(" !")
        break
```

```python
score_list = [90, 45, 70, 60, 55]

for index, score in enumerate(score_list, start=1):
    if score >= 60:
        print(f'{index}  .')
    else:
        print(f'{index}  .')
```
