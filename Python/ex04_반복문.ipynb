```python
#1. 랜덤한 수 추출(1~50) -> 변수담기(ran)
#2. 숫자 입력
#3. 입력한 숫자와 랜덤수가 같은지 큰지 작은지 비교하여 print() 출력 -> 조건문
    # 랜덤수와 입력수가 같으면 반복문 종료(break)
```

```python
import random

ran = random.randint(1, 10)

while True:
    num = int(input("숫자를 입력하세요 >> "))
    
    if num > ran:
        print(f"{num}보다 작은 수 입니다.")
    elif num < ran:
        print(f"{num}보다 큰 수 입니다.")
    else:
        print("정답을 맞추셨습니다!")
        break
```

```python
score_list = [90, 45, 70, 60, 55]

for index, score in enumerate(score_list, start=1):
    if score >= 60:
        print(f'{index}번 학생은 합격입니다.')
    else:
        print(f'{index}번 학생은 불합격입니다.')
```

```python
score_list = [90, 45, 70, 60, 55]

index = 1
for i in score_list:
    if i > 60:
        print(f'{index}번 학생은 합격입니다.')
    else:
        print(f'{index}번 학생은 불합격입니다.')
    index +=1
```
