### 1. 리스트 값 추가(append)

```python
list2 = [0,1,2,4,5]
# 리스트 가장 뒤에 값을 하나씩 추가하는 함수
# 리스트명.append(값)
list2.append(6)
```

```python
# 리스트 안에 또 다른 리스트 넣기 가능!
list2.append([0,1,2])
list2
```

```python
f_list = ['사과','포도',['수박','멜론'],'복숭아','딸기','오렌지']
```

```python
choice_list = []
choice_list.append(f_list[0])
choice_list.append(f_list[2][0])
choice_list.append(f_list[5])
```

```python
choice_list
```

### 2. 원하는 인덱스 위치에 값을 추가(insert)

```python
# 리스트명.insert(인덱스번호, 값)
# 자동으로 인덱스가 밀림
# choice_list 사과와 수박 사이에 포도 넣기
choice_list.insert(1, '포도')
choice_list
```

```python
music_list = [['Hype boy', 'NewJeans'], ['LOVE DIVE', '아이브'], ['사건의 지평선', '윤하']]

loc = int(input('인덱스 위치 입력'))
song = input('노래제목 입력 >> ')
singer = input('가수 입력 >> ')

music_list.insert(loc, [song, singer])

print(music_list)
```
