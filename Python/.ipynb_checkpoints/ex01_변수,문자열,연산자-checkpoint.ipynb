### 문자열 인덱싱 슬라이싱
- 문자열도 인덱스 번호를 가진다
- 인덱싱 : 인덱스 번호 하나를 가리켜서 데이터 추출
- 슬라이싱 : 범위를 지정하여 여러개의 데이터를 추출

```python
name = "My name is YH"
```

```python
# 슬라이싱 방법: 변수명[범위]
# 변수명[시작인덱스 : 끝인덱스 + 1 ]
# 주의1: 슬라이싱에서 범위를 작성할때 끝인덱스범위는 내가 가져고오싶은 인덱스 + 1

# My
print(name[:2]) # 처음부터 가져오고 싶다면? 시작인덱스 비워두기
# name
print(name[3:7])
# YH
print(name[-2:]) # 끝까지 가져오고 싶다면? 끝인덱스 비워두기
# 처음부터 끝까지 전체를 가져오고 싶으면?
print(name[:])
```

### 문자열 포매팅

```python
month = 6
day = 25
today = "오늘은 {}월 {}일 입니다!".format(month, day)
print(today)
```

```python
month = 6
day = 25
today = f"오늘은 {month}월 {day}일 입니다!"
print(today)
```

```python
x = 100
y = 200
add = x + y
add1 = f"{x}와 {y}의 합은 {add}입니다."
print(add1)
```

### 연산자

```python
num1 = int(input("정수를 입력하세요: "))
num2 = int(input("정수를 입력하세요: "))
print(num1 + num2)
print(num1 - num2)
print(num1 * num2)
print(num1 / num2)
```
