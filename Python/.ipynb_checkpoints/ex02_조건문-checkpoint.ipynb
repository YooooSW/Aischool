```python
print('''
================================================
================----  ----================
================================================
== 1. == 2. ==  3.  =======insert===
==--------==----------==--------=======------===
==   600  ==    800   ==  1000  ======= O00===
=======================================------===
================================================
==++++++++++++++++++++++++++++++++++++++++++++==
==++++++++++++++++++++++++++++++++++++++++++++==
==++++++++++++++++++++++++++++++++++++++++++++==
==+++++++++++++++              +++++++++++++++==
==+++++++++++++++      AIS     +++++++++++++++==
==+++++++++++++++              +++++++++++++++==
==++++++++++++++++++++++++++++++++++++++++++++==
==++++++++++++++++++++++++++++++++++++++++++++==
================================================
================================================
  ===                                      ===
  ===                                      ===
''')

money = int(input("  >> "))

choice = int(input("  >> "))

cola = 600
cider = 800
water = 1000

if choice == 1:
    price = cola
elif choice == 2:
    price = cider
elif choice == 3:
    price = water
else:
    print(" .")
    price = 0

if money >= price and price != 0:
    change = money - price
    coins_1000 = change // 1000
    change %= 1000
    coins_500 = change // 500
    change %= 500
    coins_100 = change // 100
    
    print(f": {coins_1000}, : {coins_500}, : {coins_100}")
elif price != 0:
    print(" .")
```

```python
def display_menu():
    print('''
================================================
================----  ----================
================================================
== 1. == 2. ==  3.  =======insert===
==--------==----------==--------=======------===
==   600  ==    800   ==  1000  ======= O00===
=======================================------===
================================================
==++++++++++++++++++++++++++++++++++++++++++++==
==++++++++++++++++++++++++++++++++++++++++++++==
==++++++++++++++++++++++++++++++++++++++++++++==
==+++++++++++++++              +++++++++++++++==
==+++++++++++++++      AIS     +++++++++++++++==
==+++++++++++++++              +++++++++++++++==
==++++++++++++++++++++++++++++++++++++++++++++==
==++++++++++++++++++++++++++++++++++++++++++++==
================================================
================================================
  ===                                      ===
  ===                                      ===
''')

def insert_money():
    money = int(input("  >> "))
    return money

def select_menu():
    choice = int(input("  >> "))
    return choice

def calculate_change(money, price):
    change = money - price
    return change

def display_change(change):
    coins_1000 = change // 1000
    change %= 1000
    coins_500 = change // 500
    change %= 500
    coins_100 = change // 100
    
    print(f": {coins_1000}, : {coins_500}, : {coins_100}")

def vending_machine():
    display_menu()
    money = insert_money()
    choice = select_menu()

    prices = {1: 600, 2: 800, 3: 1000}
    if choice in prices:
        price = prices[choice]
        if money >= price:
            change = calculate_change(money, price)
            display_change(change)
        else:
            print(" .")
    else:
        print(" .")

if __name__ == "__main__":
    vending_machine()
```
