```python
score = int(input("점수입력"))
if score >=90:
    grade = "A"
elif score >=80:
    grade = "B"
elif score >=70:
    grade = "C"
elif score >=60:
    grade = "D"
else:
    grade = "F"
print(f"{score}점은 {grade}학점 입니다.")
```

```python
print('''
================================================
================---- 자판기 ----================
================================================
== 1.콜라 == 2.사이다 ==  3.물  =======insert===
==--------==----------==--------=======------===
==   600  ==    800   ==  1000  ======= O00원===
=======================================------===
================================================
==++++++++++++++++++++++++++++++++++++++++++++==
==++++++++++++++++++++++++++++++++++++++++++++==
==++++++++++++++++++++++++++++++++++++++++++++==
==+++++++++++++++              +++++++++++++++==
==+++++++++++++++      AIS     +++++++++++++++==
==+++++++++++++++              +++++++++++++++==
==++++++++++++++++++++++++++++++++++++++++++++==
==++++++++++++++++++++++++++++++++++++++++++++==
================================================
================================================
  ===                                      ===
  ===                                      ===
''')

money = int(input("돈 입력 >> "))

choice = int(input("메뉴 선택 >> "))

cola = 600
cider = 800
water = 1000

if choice == 1:
    price = cola
elif choice == 2:
    price = cider
elif choice == 3:
    price = water
else:
    print("잘못된 선택입니다.")
    price = 0

if money >= price and price != 0:
    change = money - price
    coins_1000 = change // 1000
    change %= 1000
    coins_500 = change // 500
    change %= 500
    coins_100 = change // 100
    
    print(f"천원: {coins_1000}개, 오백원: {coins_500}개, 백원: {coins_100}개")
elif price != 0:
    print("금액이 부족합니다.")
```

```python
def display_menu():
    print('''
================================================
================---- 자판기 ----================
================================================
== 1.콜라 == 2.사이다 ==  3.물  =======insert===
==--------==----------==--------=======------===
==   600  ==    800   ==  1000  ======= O00원===
=======================================------===
================================================
==++++++++++++++++++++++++++++++++++++++++++++==
==++++++++++++++++++++++++++++++++++++++++++++==
==++++++++++++++++++++++++++++++++++++++++++++==
==+++++++++++++++              +++++++++++++++==
==+++++++++++++++      AIS     +++++++++++++++==
==+++++++++++++++              +++++++++++++++==
==++++++++++++++++++++++++++++++++++++++++++++==
==++++++++++++++++++++++++++++++++++++++++++++==
================================================
================================================
  ===                                      ===
  ===                                      ===
''')

def insert_money():
    money = int(input("돈 입력 >> "))
    return money

def select_menu():
    choice = int(input("메뉴 선택 >> "))
    return choice

def calculate_change(money, price):
    change = money - price
    return change

def display_change(change):
    coins_1000 = change // 1000
    change %= 1000
    coins_500 = change // 500
    change %= 500
    coins_100 = change // 100
    
    print(f"천원: {coins_1000}개, 오백원: {coins_500}개, 백원: {coins_100}개")

def vending_machine():
    display_menu()
    money = insert_money()
    choice = select_menu()

    prices = {1: 600, 2: 800, 3: 1000}
    if choice in prices:
        price = prices[choice]
        if money >= price:
            change = calculate_change(money, price)
            display_change(change)
        else:
            print("금액이 부족합니다.")
    else:
        print("잘못된 선택입니다.")

if __name__ == "__main__":
    vending_machine()
```
